# Find Distance 2D
## Write a function find_distance_2d(x1, y1, x2, y2) that returns the distance between the two points using pythagorean formula.

```python
def find_2d_distance(x1,y1,x2,y2):
    base = x2-x1
    height = y2-y1
    distance = (base*base + height*height)**0.5
    return distance

print(f"distance: {find_2d_distance(0, 0, 3, 4)}")
print(f"distance: {find_2d_distance(1, 2, 4, 6)}")
```

# Find Distance 3D


We can create two triangles: one using x,y coordintates to derive hypoteneous of triangle A. Using hypoteneous of triangle A as base for triangle B and z coordinates as height for triangle B to derive hypoteneous for triangle B, which is the distance of point A to B.


```python
def calculate_distance_3d(x1, y1, z1, x2, y2, z2):
    import math
    t1b = x2-x1 # t1b is base of triangle 1
    t1h = y2-y1 # t1h is height of triangle 1
    t1hypo = math.sqrt(t1b*t1b + t1h*t1h) # t1hypo is hypoteneous of triangle 1, which will be used as base for triangle 2
    t2b = t1hypo # t2b is base of triangle 2
    t2h = z2-z1 # t2h is height of triangle 2
    t2hypo = math.sqrt(t2b*t2b + t2h*t2h)
    distance = t2hypo
    return distance
    
print(f"distance in scenario A : {calculate_distance_3d(0, 0, 0, 1, 1, 1)}")
print(f"distance in scenario B : {calculate_distance_3d(2, 3, 5, 2, 3, 5)}")
print(f"distance in scenario C : {calculate_distance_3d(1, 2, 3, 4, 6, 9)}")
```

Given two points, write a function fit() to learn slope and intercept of the line going through them

**m** is the slope of the line. **m = (y2-y1)/(x2-x1)**

**c** is the intercept where the line crosses the y-axis. The value of y when x = 0. **c=y1-mx1**

```python
def fit(x1, y1, x2, y2):
    if x1 != x2:
        m = (y2-y1)/(x2-x1)
        c = y1-m*x1
        return m,c
    else:
        m = 0 # If x1 is equal to x2, then the line will be a vertical line with no slope.
        c = y1-m*x1
        return m,c
```

# Find First Derivative of a Function that takes only one argument

- To derive derivative of function, we can use 'forward difference method', where h is a very small number.



<span style="background-color:yellow;">**Q.9a Write a function approx_derivative(func, x, h)**.</span>

   - 'func' accepts single argument,
   - A value x where we want to find the derivative,
   - And a small number h (like 0.0001),
   - Returns the approximate derivative of func at x.

```python
def square(x):
    x = x
    return x * x

def cube(x):
    x = x
    return x * x * x

def approx_derivative(funct, x, h):
    derivative = (funct(x+h) - funct(x))/h
    return derivative
```

```python
print(approx_derivative(square, 3, 0.01))
print("It shows that if we increase length of square by 0.01, the area of the square will increase by 0.0601 (9.0601 - 9). And, derivative will be 0.0601/0.01 ~ 6.0 (change in area/change in length)")
```

Write a function closer_point(p, a, b) that takes:

p: a tuple representing the coordinates of point P
a: a tuple representing the coordinates of point A
b: a tuple representing the coordinates of point B
The function should return:

'A' if P is closer to A
'B' if P is closer to B
'Equal' if the distances are the same

```python
def closer_point(p, a, b):
    distanceptoa = (p[0] - a[0])**2 + (p[1] - a[1])**2
    distanceptob = (p[0] - b[0])**2 + (p[1] - b[1])**2
    if distanceptoa < distanceptob:
        return "A"
    elif distanceptoa > distanceptob:
        return "B"
    else:
        return "Equal"
    
```

# Is Point P on Line in 1D?

<span style="background-color:yellow;">**Q.12 Write a function is_point_on_line_1d(a, b, p) that returns True if point p lies on the line segment between a and b, and False otherwise.**.</span>

```python
def is_point_on_line_1d(a, b, p):
    if max(a,b) >= p and min(a,b) <= p:
        return True
    else:
        return False
```

```python
print(is_point_on_line_1d(2, 5, 3))      # Output: True  
print(is_point_on_line_1d(5, 2, 3))      # Output: True  
print(is_point_on_line_1d(2, 5, 6))      # Output: False  
print(is_point_on_line_1d(4, 4, 4))      # Output: True  (A single point segment)
print(is_point_on_line_1d(4, 4, 5))      # Output: False

# With negative numbers:
print(is_point_on_line_1d(-5, -2, -3))   # Output: True  
print(is_point_on_line_1d(-5, -2, -6))   # Output: False  
print(is_point_on_line_1d(-2, -5, -4))   # Output: True  
print(is_point_on_line_1d(-1, 3, 0))     # Output: True
```

# Are Two Lines Overlapping or Touching in 1D

- Write a function are_lines_touching_or_overlapping(start1, end1, start2, end2) that returns True if the two 1D line segments are overlapping or touching, and False if they are completely separate.

📌 Make sure your function works correctly even if the start is greater than the end — the order shouldn't matter.

```python
def are_lines_touching_or_overlapping(start1, end1, start2, end2):
    if max(start2, end2) >= min(start1, end1) and min(start2, end2) <= max(start1, end1):
        return True
    else:
        return False
```

# Identify Point Inside or On a Rectangle or any other Shape (2D)?

🧠 Explanation:
In geometry, a rectangle is a shape with four sides and four right angles. When the sides of a rectangle are parallel to the x and y axes, it means the edges of the rectangle are either horizontal or vertical—not slanted.

To describe such a rectangle, we only need two opposite corners:

The bottom-left corner (x1, y1)
The top-right corner (x2, y2)
A point has two values: x and y—its horizontal and vertical positions.

To check if a point lies inside (or on the border of) the rectangle, we see if:

The x value of the point is between x1 and x2, and
The y value of the point is between y1 and y2.
This assumes x1 < x2 and y1 < y2 (which means the first point is bottom-left and the second is top-right).

<span style="background-color:yellow;">**Q.14a Is a Point Inside a Rectangle (with Sides Parallel to the Axes)?**.</span>

**Write a function is_point_inside_rectangle(x1, y1, x2, y2, px, py) that returns True if the point (px, py) lies inside or on the boundary of the rectangle defined by corners (x1, y1) and (x2, y2), and False otherwise.**

```python
def is_point_inside_rectangle(x1, y1, x2, y2, px, py):
    if px >= x1 and px <= x2:
        if py >= y1 and py <= y2:
            return True
    else:
        return False
```

```python
print(is_point_inside_rectangle(0, 0, 10, 5, 3, 2))   # Output: True
print(is_point_inside_rectangle(0, 0, 10, 5, 10, 5))  # Output: True  (point on the corner)
print(is_point_inside_rectangle(0, 0, 10, 5, 11, 5))  # Output: False (outside the rectangle)
print(is_point_inside_rectangle(-5, -5, 5, 5, 0, 0))  # Output: True  (inside a rectangle with negative coordinates)
```

<span style="background-color:yellow;">**Q.14b Are rectangles (with sides parallel to axes) intersecting?**.</span>

Two rectangles are said to intersect if they share any area — even a single point on their boundary counts. If one is completely to the left, right, above, or below the other, then they do not intersect.

**Write a function are_rectangles_intersecting(rect1, rect2) that takes two rectangles and returns True if they intersect, otherwise returns False.**

Each rectangle is represented as a tuple of two points: ((x1, y1), (x2, y2)), where



(x1, y1) is the bottom-left corner
(x2, y2) is the top-right corner

```python
def are_rectangles_intersecting(rect1, rect2):
    (rect1_x1, rect1_y1), (rect1_x2, rect1_y2) = rect1
    (rect2_x1, rect2_y1), (rect2_x2, rect2_y2) = rect2
    if (rect2_x1 >= rect1_x1 and rect2_x1 <= rect1_x2) and (rect2_y1 >= rect1_y1 and rect2_y1 <= rect1_y2):
        return True
    else:
        return False
    
```

```python
# Rectangles overlap partially
print(are_rectangles_intersecting(((0, 0), (3, 3)), ((2, 2), (5, 5))))  
# Output: True

# One rectangle is completely to the right of the other
print(are_rectangles_intersecting(((0, 0), (1, 1)), ((2, 2), (3, 3))))  
# Output: False

# Touching at corner
print(are_rectangles_intersecting(((0, 0), (2, 2)), ((2, 2), (4, 4))))  
# Output: True

# One rectangle inside another
print(are_rectangles_intersecting(((0, 0), (5, 5)), ((1, 1), (2, 2))))  
# Output: True
```

## Factorial using recursion

### **Write a function factorial_recursive(n) that takes one argument n (a non-negative integer) and returns the factorial of that number using recursion.
If n is 0, the function should return 1.**

```python
# if-else without recursion
def factorial_without_recursion(n):
    num = int(n)
    if n < 0:
        return "Negative number"
    elif type(n) != int:
        return "Invalid number"
    else:
        fact = 1
        for i in range(1,num+1):
            fact = fact*i
#             print(i, fact)
        return fact
        
```

```python
# if-else with recrusion
def factorial_recursive(n):
    # print(f"value of n in each iteration of if-else loop: {n}")
    if n < 0:
        return "Negative number"
    elif n == 0 or n == 1: #BASE CASE - recursion ends when n == 1 because that’s when the function stops calling itself.
        return 1
    else:
        return n*factorial_recursive(n-1)
```

```python
print(factorial_recursive(5))  # Output: 120  
print(factorial_recursive(3))  # Output: 6  
print(factorial_recursive(0))  # Output: 1  
print(factorial_recursive(1))  # Output: 1 
```

## Compute HCF using Euclid's Method with Recursion



### Write a function named compute_hcf(a, b) that takes two positive integers a and b and returns their HCF using Euclid's method with recursion.

You should use recursion to solve this problem.
The function should return an integer.

```python
def compute_hcf(a, b):
    if a == 0 or b == 0: # Base case of recursion
        return max(a,b)
    else:
        return compute_hcf(b, a%b) # hcf(a, b) = hcf(b, a % b)
```

```python
print(compute_hcf(12, 18))     # Output: 6
print(compute_hcf(100, 25))    # Output: 25
print(compute_hcf(17, 13))     # Output: 1
print(compute_hcf(0, 5))       # Output: 5
```

```python
def multiply_recursive(a, b):
    org_a = a
    org_b = b
    a, b = abs(a), abs(b)
    if b == 0:
        return 0
    else:
        num = a + multiply_recursive(a, b-1)
#         print("b is ", b)
        if (org_a > 0 and org_b >0) or (org_a < 0 and org_b < 0): 
            return num 
        else: 
            return -num
```

```python
def multiply_recursive_approach2(a,b):
    if b == 0:
        return 0
    elif b > 0:
        return a + multiply_recursive_approach2(a,b-1)
    else:
        return -multiply_recursive_approach2(a,-b)
```

## Division using recursion to find quotient and remainder



###  Write a function recursive_divide(dividend, divisor) that returns a tuple (quotient, remainder) using recursion. You must not use the // or % operators.

dividend: a non-negative integer

divisor: a positive integer

Return: A tuple of two integers: (quotient, remainder)

```python
def recursive_divide(dividend, divisor):
#     num = ()
    if dividend < divisor:
        return (0, dividend)
    else:
        quotient, remainder = recursive_divide(dividend - divisor, divisor)
        return quotient+1, remainder
```

```python
print(recursive_divide(17, 5))   # Output: (3, 2)
print(recursive_divide(20, 4))   # Output: (5, 0)
print(recursive_divide(7, 3))    # Output: (2, 1)
print(recursive_divide(0, 1))    # Output: (0, 0)
```
